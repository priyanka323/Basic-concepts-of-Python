```python
class myClass:
  def sum(self):
    int_res=100+200
    return int_res
class newClass(myClass):
  def hello(self):
    print("maya")
classObj=newClass()
classObj.hello()
int_sum=classObj.sum()
print("the sum is:",int_sum)      


```
