```python
# 0 1 1 2 3 5 8 13
def fibseq(n):
  x = 0
  y = 1
  if n ==1:
    print(x)
  else:
    print(x,end=" ")
    print(y,end=' ')

  for i in range(2, n):
    c  = x+y
    x = y
    y = c
    print(c,end="")
fibseq(6)
```
